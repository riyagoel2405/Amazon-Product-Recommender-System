```python
import os
import pandas as pd
import kagglehub

path = kagglehub.dataset_download("arhamrumi/amazon-product-reviews")
print("Path to dataset files:", path)

# inspect files
files = os.listdir(path)
print(files)  # see actual CSV filename here

# then pick the right one
csv_path = os.path.join(path, 'Reviews.csv')  # or whatever name you saw above
df = pd.read_csv(csv_path)

print("rows, cols:", df.shape)
print(df.columns.tolist())
df.head()
```

```python
import pandas as pd

if 'HelpfulnessDenomenator' in df.columns:
    df = df.rename(columns={'HelpfulnessDenomenator': 'HelpfulnessDenominator'})

ratings = df[['UserId', 'ProductId', 'Score']].dropna().copy()
ratings.columns = ['user_id', 'item_id', 'rating']
ratings['rating'] = ratings['rating'].astype(float)

# Filter: keep users with ≥ 5 ratings and items with ≥ 5 ratings (tuneable)
min_user_ratings = 5
min_item_ratings = 5

user_counts = ratings['user_id'].value_counts()
item_counts = ratings['item_id'].value_counts()
ratings = ratings[ratings['user_id'].isin(user_counts[user_counts >= min_user_ratings].index)]
ratings = ratings[ratings['item_id'].isin(item_counts[item_counts >= min_item_ratings].index)]

print("after filtering:", ratings.shape)
```

```python
from surprise import Dataset, Reader

reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings[['user_id', 'item_id', 'rating']], reader)
```
